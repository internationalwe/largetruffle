```python
# 한글 폰트 사용을 위해서 세팅
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

# 폰트 사이즈 조정
SMALL_SIZE = 12
MEDIUM_SIZE = 14 
BIGGER_SIZE = 20
plt.rc('font', size=SMALL_SIZE) # controls default text sizes 
plt.rc('axes', titlesize=BIGGER_SIZE) # fontsize of the axes title 
plt.rc('axes', labelsize=MEDIUM_SIZE) # fontsize of the x and y labels 
plt.rc('xtick', labelsize=SMALL_SIZE) # fontsize of the tick labels 
plt.rc('ytick', labelsize=SMALL_SIZE) # fontsize of the tick labels 
plt.rc('legend', fontsize=SMALL_SIZE) # legend fontsize 
plt.rc('figure', titlesize=BIGGER_SIZE) # fontsize of the figure title



# 막대그래프
x = np.arange(4)
years = ['수납장', '의자', '책상', '침대'] #['Closet', 'Chair', 'Desk', 'Bed'] #x축 
values = [216, 334, 230, 83] #y축


plt.title('< 카테고리별 갯수 >')
plt.xlabel('카테고리')
plt.ylabel('갯수')
plt.bar(x, values, color = ['red', 'green', 'blue', 'black'])
plt.xticks(x, years)
plt.show()
```

수평 막대그래프

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.arange(4)
years = ['Closet', 'Chair', 'Desk', 'Bed'] #y축
values = [216, 334, 230, 83] #x축

plt.barh(y, values, height=-0.6, align='edge',
        edgecolor="gray", linewidth=3, tick_label=years, log=False, color = ['red', 'green', 'blue', 'black'])
plt.ylabel('카테고리')
plt.xlabel('갯수')
plt.show()
```

파이 차트

```python
import matplotlib.pyplot as plt

ratio = [20, 20, 60]
labels = ['Test', 'Validation', 'Train']
colors = ['#ff9999', '#ffc000', '#8fd9b6']
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, colors=colors, wedgeprops=wedgeprops)
plt.show()
```
